```python
# Instalar a biblioteca que gera QR Code

from google.colab import files
import qrcode
from matplotlib import pyplot as plt

# Gerar e salvar o primeiro QR Code
imagem = qrcode.make("https://ciia-saude.medicina.ufmg.br/")

# Salvar a imagem do QR Code
imagem.save("qrCode_GitJulianoGaspar2.png")

# Exibir a imagem para verificação
plt.imshow(imagem)
plt.axis('off')  # Desligar os eixos
plt.show()

# Fazer o download da imagem gerada
files.download("qrCode_GitJulianoGaspar2.png")
```

```python
# Gerar e salvar o segundo QR Code (endereço)
imagem = qrcode.make('Av. Alfredo Balena, 190, Centro, BH')

# Salvar a imagem do QR Code
imagem.save("enderecoFM.jpg")

# Exibir a imagem para verificação
plt.imshow(imagem)
plt.axis('off')  # Desligar os eixos
plt.show()

# Fazer o download da imagem gerada
files.download("enderecoFM.jpg")
```

```python
# Gerar um QR Code colorido
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_L,
    box_size=10,
    border=4
)

# Adicionar dados ao QR Code
qr.add_data('https://ciia-saude.medicina.ufmg.br')
qr.make(fit=True)

# Criar a imagem com cores personalizadas
img = qr.make_image(fill_color=(139, 195, 74), back_color=(0, 34, 92))

# Salvar o QR Code colorido
img.save("qrCIIASAUDE_azul.png")

# Exibir a imagem para verificação
plt.imshow(img)
plt.axis('off')  # Desligar os eixos
plt.show()

# Fazer o download da imagem gerada
files.download("qrCIIASAUDE_azul.png")
```
